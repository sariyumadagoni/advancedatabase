```python
import pandas as pd

# Query 1 (6.1.4 a)
# Find the class name and country for all classes with at least 10 guns.
query1 = """
SELECT class, country
FROM Classes
WHERE numGuns >= 10;
"""
cur.execute(query1)
rows = cur.fetchall()
df1 = pd.DataFrame(rows, columns=["class","country"])
print("Query 1 Results:")
display(df1)
```

```python
# Query 2 (6.1.4 b)
# Find the names of all ships launched prior to 1918 (rename column to shipName).
query2 = """
SELECT name AS shipName
FROM Ships
WHERE launched < 1918;
"""
cur.execute(query2)
rows = cur.fetchall()
df2 = pd.DataFrame(rows, columns=["shipName"])
print(" Query 2 Results:")
display(df2)
```

```python
# Query 3 (6.2.3 a)
# Find the ships heavier than 35,000 tons.
# ------------------------------
query3 = """
SELECT S.name, C.displacement
FROM Ships S
JOIN Classes C ON S.class = C.class
WHERE displacement > 35000;
"""
cur.execute(query3)
rows = cur.fetchall()
df3 = pd.DataFrame(rows, columns=["ship","displacement"])
print("Query 3 Results:")
display(df3)
```
